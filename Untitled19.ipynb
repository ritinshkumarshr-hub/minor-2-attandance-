```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df['Average_Score'] = (df['math score'] + df['reading score'] + df['writing score']) / 3
```

```python
def categorize_score(score):
    if score >= 80:
        return "Excellent"
    elif score >= 60:
        return "Good"
    elif score >= 40:
        return "Average"
    else:
        return "Poor"

df['Score_Category'] = df['Average_Score'].apply(categorize_score)
df.head()
```

```python
average_scores_by_gender = df.groupby('gender')['Average_Score'].mean().reset_index()

plt.figure(figsize=(8, 6))
plt.bar(average_scores_by_gender['gender'], average_scores_by_gender['Average_Score'], color=['skyblue', 'lightcoral'])
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.title("Average Scores by Gender")
plt.show()
```

```python
score_category_counts = df['Score_Category'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(score_category_counts, labels=score_category_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Score Category Distribution")
plt.show()
```
